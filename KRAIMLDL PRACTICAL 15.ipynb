```python
df=pd.DataFrame({'filename':filenames})
df.head()
```

```python
df['class']=df['filename'].apply(lambda X:X[:3])
```

```python
df.head()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_gen=ImageDataGenerator(zoom_range=0.2,shear_range=0.2,horizontal_flip=True,rescale=1/255)
```

```python
train_data=data_gen.flow_from_dataframe(df,'C:/Users/akmeh/Music/KRAI Assignments/train',X='filename',y='class',target_size=(224,224))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense
```

```python
model=Sequential()
model.add(Conv2D(16,(3,3),activation='relu',input_shape=(224,224,3)))
model.add(MaxPool2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Conv2D(64,(5,5),activation='relu'))
model.add(MaxPool2D())
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Flatten())
model.add(Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_data,epochs=5)
```

```python
import cv2
def get_class(img_path):
    img=cv2.imread(img_path)
    img=cv2.resize(img,(224,224))
    img=img/255
    op=model.predict(img.reshape(1,224,224,3)).argmax()
    return 'cat' if op==0 else 'dog'
```
